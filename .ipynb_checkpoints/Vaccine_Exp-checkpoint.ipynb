```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
# Parameters for SEIR Model
N = 1e6          # Total population
beta = 0.3       # Infection rate
gamma = 1/14     # Recovery rate (1/recovery_days)
sigma = 1/5      # Rate of progression from exposed to infected
initial_infected = 10
initial_exposed = 5
initial_recovered = 0
initial_susceptible = N - initial_infected - initial_exposed

# SEIR model differential equations
def deriv(y, t, N, beta, sigma, gamma):
    S, E, I, R = y
    dSdt = -beta * S * I / N
    dEdt = beta * S * I / N - sigma * E
    dIdt = sigma * E - gamma * I
    dRdt = gamma * I
    return dSdt, dEdt, dIdt, dRdt

# Initial conditions vector
y0 = [initial_susceptible, initial_exposed, initial_infected, initial_recovered]

# Time points (0 to 15 weeks, daily resolution)
t = np.linspace(0, 15 * 7, 15 * 7)

# Integrating the SEIR equations
ret = odeint(deriv, y0, t, args=(N, beta, sigma, gamma))
S, E, I, R = ret.T
```

```python
# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(t / 7, I, label='Infected')
plt.plot(t / 7, S, label='Susceptible')
plt.plot(t / 7, E, label='Exposed')
plt.plot(t / 7, R, label='Recovered')
plt.xlabel('Weeks')
plt.ylabel('Population')
plt.title('SEIR Model (Unvaccinated Period)')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
# Example data: weekly cases (dummy data for unvaccinated phase)
weekly_cases = [100, 150, 200, 300, 500, 800, 1300, 2100, 3400, 5500, 8900, 14400, 23200, 37400, 60200]

# Fit ARIMA model
model = ARIMA(weekly_cases, order=(2, 1, 2))
model_fit = model.fit()

# Forecast next 10 weeks
forecast = model_fit.forecast(steps=10)
```

```python
# Plot the forecast
plt.figure(figsize=(10, 6))
plt.plot(range(len(weekly_cases)), weekly_cases, label='Observed')
plt.plot(range(len(weekly_cases), len(weekly_cases) + 10), forecast, label='Forecast')
plt.xlabel('Weeks')
plt.ylabel('Cases')
plt.title('ARIMA Model Forecast (Unvaccinated Period)')
plt.legend()
plt.show()
```

```python
# Parameters for vaccinated compartments
vaccination_start = 16  # Week vaccination starts
one_dose_efficacy = 0.6
two_dose_efficacy = 0.9
vaccination_rate = 50000  # People vaccinated per week

# Extend the SEIR model to include vaccination
def deriv_with_vaccine(y, t, N, beta, sigma, gamma, one_dose_efficacy, two_dose_efficacy, vaccination_rate):
    S, E, I, R, V1, V2 = y
    dSdt = -beta * S * I / N - vaccination_rate
    dEdt = beta * S * I / N - sigma * E
    dIdt = sigma * E - gamma * I
    dRdt = gamma * I
    dV1dt = vaccination_rate - one_dose_efficacy * beta * V1 * I / N  # One dose reduces transmission by efficacy
    dV2dt = one_dose_efficacy * beta * V1 * I / N - two_dose_efficacy * beta * V2 * I / N  # Two-dose protection
    return dSdt, dEdt, dIdt, dRdt, dV1dt, dV2dt

# Initial conditions vector with vaccinated compartments
y0_vaccine = [initial_susceptible, initial_exposed, initial_infected, initial_recovered, 0, 0]

# Time points (16 to 40 weeks)
t_vaccine = np.linspace(16 * 7, 40 * 7, 24 * 7)

# Integrating the SEIR equations with vaccination
ret_vaccine = odeint(deriv_with_vaccine, y0_vaccine, t_vaccine, args=(N, beta, sigma, gamma, one_dose_efficacy, two_dose_efficacy, vaccination_rate))
S_v, E_v, I_v, R_v, V1_v, V2_v = ret_vaccine.T
```

```python
# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(t_vaccine / 7, I_v, label='Infected with Vaccination')
plt.plot(t_vaccine / 7, V1_v, label='One Dose Vaccinated')
plt.plot(t_vaccine / 7, V2_v, label='Two Dose Vaccinated')
plt.xlabel('Weeks')
plt.ylabel('Population')
plt.title('SEIR Model with Vaccination (Post-Vaccination Phase)')
plt.legend()
plt.show()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
```

```python
# Truncate arrays to the minimum length to avoid mismatch
min_length = min(len(I_v), len(V1_v), len(V2_v), len(np.arange(16, 40)))

# Prepare dataset: weekly cases, vaccination rates, etc.
data = pd.DataFrame({
    'week': np.arange(16, 40)[:min_length],        # Truncate to min_length
    'infected_unvaccinated': I_v[:min_length],     # Truncate to min_length (infected cases)
    'one_dose_vaccinated': V1_v[:min_length],      # Truncate to min_length (one-dose vaccinated cases)
    'two_dose_vaccinated': V2_v[:min_length]       # Truncate to min_length (two-dose vaccinated cases)
})

# Display the resulting DataFrame
data.head()
```

## Unvaccinated

```python
# Normalizing data for LSTM (unvaccinated)
scaler_unvacc = MinMaxScaler(feature_range=(0, 1))
scaled_data_unvacc = scaler_unvacc.fit_transform(data[['infected_unvaccinated']])

# Prepare sequences for LSTM (X: features, y: infected)
X_unvacc = []
y_unvacc = []
sequence_length = 4
for i in range(sequence_length, len(scaled_data_unvacc)):
    X_unvacc.append(scaled_data_unvacc[i-sequence_length:i])
    y_unvacc.append(scaled_data_unvacc[i, 0])

X_unvacc, y_unvacc = np.array(X_unvacc), np.array(y_unvacc)

# LSTM Model for unvaccinated population
model_unvacc = Sequential()
model_unvacc.add(LSTM(50, return_sequences=True, input_shape=(X_unvacc.shape[1], X_unvacc.shape[2])))
model_unvacc.add(LSTM(50))
model_unvacc.add(Dense(1))

model_unvacc.compile(optimizer='adam', loss='mean_squared_error')
model_unvacc.fit(X_unvacc, y_unvacc, epochs=100, batch_size=32)
```

```python
# Make predictions for unvaccinated
predictions_unvacc = model_unvacc.predict(X_unvacc)

# Inverse transform predictions to original scale
predicted_cases_unvacc = scaler_unvacc.inverse_transform(np.concatenate([predictions_unvacc], axis=1))
```

```python
# Plot results
plt.plot(data['week'][sequence_length:], data['infected_unvaccinated'][sequence_length:], label='Actual Unvaccinated')
plt.plot(data['week'][sequence_length:], predicted_cases_unvacc, label='Predicted Unvaccinated')
plt.legend()
plt.title('Infected Cases - Unvaccinated Population')
plt.show()
```

```python
# Truncate arrays to the minimum length to avoid mismatch
min_length = min(len(I_v), len(V1_v), len(V2_v), len(np.arange(16, 40)))

# Prepare dataset: weekly cases, vaccination rates, etc.
data = pd.DataFrame({
    'week': np.arange(16, 40)[:min_length],  # Truncate to min_length
    'infected': I_v[:min_length],            # Truncate to min_length
    'one_dose_vaccinated': V1_v[:min_length],  # Truncate to min_length
    'two_dose_vaccinated': V2_v[:min_length]   # Truncate to min_length
})

# Normalizing data for LSTM
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data[['infected', 'one_dose_vaccinated', 'two_dose_vaccinated']])

# Prepare sequences for LSTM (X: features, y: infected)
X = []
y = []
sequence_length = 4
for i in range(sequence_length, len(scaled_data)):
    X.append(scaled_data[i-sequence_length:i])
    y.append(scaled_data[i, 0])

X, y = np.array(X), np.array(y)
```

```python
# LSTM Model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(X.shape[1], X.shape[2])))
model.add(LSTM(50))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X, y, epochs=100, batch_size=32)
```

```python
# Make predictions
predictions = model.predict(X)

# Inverse transform predictions to original scale
predicted_cases = scaler.inverse_transform(np.concatenate([predictions, np.zeros((len(predictions), 2))], axis=1))[:, 0]
```

```python
# Plot results
plt.figure(figsize=(10, 6))
plt.plot(data['week'][sequence_length:], data['infected'][sequence_length:], label='Actual')
plt.plot(data['week'][sequence_length:], predicted_cases, label='Predicted')
plt.xlabel('Weeks')
plt.ylabel('Infected Cases')
plt.title('LSTM Model Prediction (Post-Vaccination)')
plt.legend()
plt.savefig('LSTM Model Prediction (Post-Vaccination)')
plt.show()
```
